**DATA VISUALIZATION**

```python
#  boxplot mean of age
sns.boxplot(x=data['Age'],palette='BuGn')
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='Attrition', hue='JobSatisfaction', data=data, palette='viridis')
```

```python
# Average ages for each job compared to males and females
plt.figure(figsize = (22, 12))
sns.boxplot(x = 'JobRole', y = 'Age', hue = 'Gender', data = data, palette = "BuGn")
```

```python
plt.figure(figsize=(20, 10))
sns.boxplot(x='JobRole', y='Age', hue='Department', data= data, palette='YlGnBu')
plt.show()
```

```python
# The number of employees in the field of education, whether male or female
plt.figure(figsize=(10,5))
sns.countplot(x='EducationField',hue='Gender',data= data, palette='YlGn')
```

```python
 # OverTime comparison through Departments
 plt.figure(figsize=(10,5))
 sns.countplot(x='Department',hue='OverTime',data= data,palette='PuRd')
```

```python
# pieplot show MaritalStatus rate and numbers all Statu
# Here I took the age average for each marital status in order to display perce
#Also, as the data increases, each percentage changes, and there are many ways
y_age = [37.4,37.7, 35.3]
lab =data['MaritalStatus'].value_counts()
col = ['r','y','b']
plt.pie(y_age,labels=lab,colors=col,startangle=90,shadow=True,explode=(0.1,0,0))
plt.legend()
plt.show()
```
